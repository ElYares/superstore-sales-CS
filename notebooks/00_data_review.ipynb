# 00 - Data Review

This notebook validates the processed **Global Superstore** dataset before deeper exploratory analysis.

The goal is to confirm:

- Dataset dimensions.
- Column names.
- Data types.
- Missing values.
- Duplicates.
- Date ranges.
- Main business KPIs.
- Key categorical variables.
- Basic quality checks.

This notebook uses reusable code from the `superstore` package instead of duplicating logic inside the notebook.

```python
#IMPORTS
import pandas as pd
from superstore.data.loader import load_processed_data
from superstore.analytics.kpis import calculate_main_kpis
```

```python
# Notebook Settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", "{:,.4f}".format)
```

```python
# Load processed dataset

df = load_processed_data()

df.head()
```

```python
# Dataset shape

rows, columns = df.shape

print(f"Rows: {rows:,}")
print(f"Columns: {columns:,}")
```

```python
# Columns

df.columns.tolist()
```

```python
# Data types
############
# Se busca que las fechas tengan datetime, con eso podemos hacer un analicis de ventas por
# Anio, mes, trimestre, dia de la semana y tiempos de envio
############

df.dtypes
```

```python
# Missing values

missing_values = (
    df.isna()
    .sum()
    .sort_values(ascending=False)
)

missing_values[missing_values > 0]
```

```python
# Missing values summary

total_missing_values = int(df.isna().sum().sum())

print(f"Total missing values: {total_missing_values:,}")
```

```python
# Duplicate rows

duplicate_rows = int(df.duplicated().sum())
print(f"Duplicated rows: {duplicate_rows:,}")
```

```python
# Date columns validation

df[["order_date", "ship_date"]].dtypes
```

```python
# Date range

date_summary = {
    "min_order_date": df["order_date"].min(),
    "max_order_date": df["order_date"].max(),
    "min_ship_date": df["ship_date"].min(),
    "max_ship_date": df["ship_date"].max(),
}

date_summary
```

```python
# Check invalid shipping days

invalid_shipping_days = df[df["shipping_days"] < 0]

print(f"Rows with invalid shipping days: {len(invalid_shipping_days)}")
```

# Senial 1: El negocio es rentable, pero no con margen enorme

Profit margin: 11.61%

Esto significa

> Por cada $100 vendidos, quedan aproximadamente 11.61 de utilidad

No es malo, pero tampoco es un margen alto.

## Hipotesis

Pequenias malas decisiones en:

- Descuentos
- Envios
- Productos

Pueden destruir la utilidad

# Senial 2: El costo logistico es muy relevante

Shipping cost ratio: 10.70%

Esto significa:

> Por cada USD 100 vendidos, aproximadamente USD 10.70 se van en shipping cost.

Comprando con el margen de utilidad de 11.61%, logistica es critica

# Senial 3: Casi 1 de cada 4 ordenes tiene perdida

loss_order_rate: 24.46%

Esto significa:

> Aunque el negocio gana dinero en total, hay muchas ordenes que pierden dinero.

## Hipotesis o pregunta
La siguiente pregunta no es “¿vendemos mucho?”, sino:

- ¿En qué regiones?
- ¿En qué mercados?
- ¿En qué categorías?
- ¿En qué productos?
- ¿Con qué descuentos?
- ¿Con qué métodos de envío?

```python
# Formatted KPIs

formatted_kpis = {
    "Total sales": f"${kpis['total_sales']:,.2f}",
    "Total profit": f"${kpis['total_profit']:,.2f}",
    "Profit margin": f"{kpis['profit_margin']:.2%}",
    "Total orders": f"{kpis['total_orders']:,}",
    "Total customers": f"{kpis['total_customers']:,}",
    "Average order value": f"${kpis['average_order_value']:,.2f}",
    "Average profit per order": f"${kpis['average_profit_per_order']:,.2f}",
    "Average discount": f"{kpis['average_discount']:.2%}",
    "Total quantity": f"{kpis['total_quantity']:,}",
    "Total shipping cost": f"${kpis['total_shipping_cost']:,.2f}",
    "Average shipping cost": f"${kpis['average_shipping_cost']:,.2f}",
    "Shipping cost ratio": f"{kpis['shipping_cost_ratio']:.2%}",
    "Loss rows": f"{kpis['loss_rows']:,}",
    "Loss row rate": f"{kpis['loss_row_rate']:.2%}",
    "Loss orders": f"{kpis['loss_orders']:,}",
    "Loss order rate": f"{kpis['loss_order_rate']:.2%}",
}
formatted_kpis
```

```python
# Categorical columns overviews

categorical_columns = [
    "market",
    "market_group",
    "region",
    "country",
    "category",
    "sub_category",
    "segment",
    "ship_mode",
    "order_priority",
    "discount_level",
]

categorical_summary = pd.DataFrame({
    "column": categorical_columns,
    "unique_values": [
        df[column].nunique()
        for column in categorical_columns
    ],
})

categorical_summary
```

```python
# Unique values for main categorical variables
main_categorical_columns = [
    "market",
    "market_group",
    "category",
    "segment",
    "ship_mode",
    "order_priority",
    "discount_level",
]

for column in main_categorical_columns:
    print(f"\n{column}")
    print(df[column].dropna().unique())
```

```python
# Numeric variables summary

numeric_columns = [
    "sales",
    "profit",
    "quantity",
    "discount",
    "shipping_cost",
    "shipping_days",
    "profit_margin",
]

df[numeric_columns].describe()
```

```python
# Sales and profit sanity check

sanity_check = {
    "sales_less_or_equal_zero": int((df["sales"] <= 0).sum()),
    "profit_negative_rows": int((df["profit"] < 0).sum()),
    "quantity_less_or_equal_zero": int((df["quantity"] <= 0).sum()),
    "discount_less_than_zero": int((df["discount"] < 0).sum()),
    "discount_greater_than_one": int((df["discount"] > 1).sum()),
    "shipping_cost_less_than_zero": int((df["shipping_cost"] < 0).sum()),
}

sanity_check
```

```python
# Dateset period by year

orders_by_year = (
    df.groupby("order_year")
    .agg(
        total_sales=("sales", "sum"),
        total_profit=("profit", "sum"),
        total_orders=("order_id", "nunique"),
    )
    .reset_index()
)

orders_by_year
```

```python
# Basic business distribution

business_distribution = {
    "markets": df["market"].nunique(),
    "market_groups": df["market_group"].nunique(),
    "regions": df["region"].nunique(),
    "countries": df["country"].nunique(),
    "categories": df["category"].nunique(),
    "sub_categories": df["sub_category"].nunique(),
    "products": df["product_id"].nunique(),
    "customers": df["customer_id"].nunique(),
    "orders": df["order_id"].nunique(),
}

business_distribution
```
